```python
import torch
from diffusers import StableDiffusionXLPipeline
model_path = "./full_merged_model"
#full SDXL pipeline
pipe = StableDiffusionXLPipeline.from_pretrained(
    model_path,
    torch_dtype=torch.float16,
)
```

```python
from PIL import Image
from IPython.display import display
prompt = "A little prince playing in a beautiful garden, coloring book"
negative_prompt = "color, photorealistic, detailed shading, realistic lighting"
image = pipe(prompt=prompt, negative_prompt=negative_prompt, num_inference_steps=30).images[0]
display(image)
```
